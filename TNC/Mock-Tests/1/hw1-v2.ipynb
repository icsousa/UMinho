## Exercício 3
Sejam n=2^{30}-1 e m=2^{29}-1 , e considere o isomorfismo de anéis f de Znm para ZnxZm, tal como definido nas aulas. Calcule a imagem recíproca, por f, da soma de (a_1, a_2) com (b_1, b_2), tomando (a_1, a_2)= (415979923, 139734214) e (b_1, b_2)= (765743091, 165246078).

```python
n = 2**30 - 1
m = 2**29 - 1
a1, a2 = 415979923, 139734214
b1, b2 = 765743091, 165246078

s1 = (a1 + b1) % n
s2 = (a2 + b2) % m
x = crt([s1, s2], [n, m])
x % (n*m)
```
