## Exercício 5

```python
n = 2**30 - 1
m = 2**29 - 1 
a1, a2 = 586698395, 483591043
b1, b2 = 982504967, 244746282

s1 = (a1 + b1) % n
s2 = (a2 + b2) % m
x = crt([s1,s2],[n,m])
x % (n*m)
```
