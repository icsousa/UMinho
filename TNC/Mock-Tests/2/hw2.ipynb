## Exercício 1
Para n = 108659 , indique exatamente quais das seguintes bases n é pseudo primo de Euler:

```python
n = 108659
bases = [2643, 19071, 6289, 108658]

for a in bases:
    lhs = power_mod(a, (n-1)//2, n)
    jacobi = jacobi_symbol(a,n)
    rhs = n-1 if jacobi == -1 else jacobi
    if lhs == rhs:
        print(a)
```

## Exercício 3
Para p = 16921 , indique exatamente quais dos seguintes naturais são raizes primitivas de p:

```python
p = 16921
phi_p = p-1
candidatos = [8518, 5988, 3631, 9168]

Zp = IntegerModRing(p)

for c in candidatos:
    elemento = Zp(c)
    ordem = elemento.multiplicative_order()

    if ordem == phi_p:
        print(c)
```

## Exercício 4
Dada a chave pública ElGamal (p, r, b)=( 3137291693 , 2 , 424669327 ), a decifração de ( 870116626 , 366464538 ) é

```python
p = 3137291693
r = 2
b = 424669327
gama = 870116626
delta = 366464538

F = GF(p)
r_f = F(r)
b_f = F(b)
g_f = F(gama)
d_f = F(delta)

a = discrete_log(b_f, r_f)

mensagem = d_f*(g_f^(-a))
print(mensagem)
```

## Exercício 5
Dada a chave pública ElGamal (p, r, b)=( 11788657990214900999 , 7 , 5434812812956634422 ) a cifração de mens=1234, com o parâmetro aleatório k= 11202467694684713080 , é (gama, delta) =

```python
p = 11788657990214900999
r = 7
b = 5434812812956634422
mens = 1234
k = 11202467694684713080

Zp = IntegerModRing(p)

r_zp = Zp(r)
b_zp = Zp(b)
m_zp = Zp(mens)

gama = r_zp^k
delta = m_zp*(b_zp^k)

gama, delta
```
