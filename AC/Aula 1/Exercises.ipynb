### 1. Age checker
Ask for the user’s age and print:

### 2. String cleaner
Given the string `" Adenosine Triphosphate "`, do the following:  
1. Strip spaces  
2. Convert to lowercase  
3. Replace "a" with "@"

```python
def string_cleaner(string):
    result_string = []
    string_strip = string.lower().replace(' ', '')
    for c in string_strip:
        if c == "a":
            result_string.append("@")
        else:
            result_string.append(c)
    print(result_string)

string_cleaner("Adenosine Triphosphate")
```

### 3. CompScience FizzBuzz
Print numbers from 1 to 30, but:  
- Replace multiples of 3 with "Comp"  
- Replace multiples of 5 with "Science"  
- Replace multiples of both with "CompScience"

```python
def compScience_FizzBuzz():
    lista = list(range(1,31))
    for n in lista:
        if n % 3 == 0 and n % 5 == 0:
            print("CompScience")
        elif n % 3 == 0:
            print("Comp")
        elif n % 5 == 0:
            print("Science")
        else:
            print(n)

compScience_FizzBuzz()
```

### 4. Guess the number
The program generates a random number between 1 and 20.  
The user must guess until correct.  
Print hints "too high" or "too low".

```python
import random

def guess():
    r = random.randint(1, 20)
    c = 1
    while c:
        n = int(input("Input your lucky number: "))
        if n < r:
            print("Too Low")
        elif n > r:
            print("Too High")
        else:
            c = 0
            print("You Win")

guess()
```

### 5. Factorial
Ask the user for a number `n` and compute `n!` (factorial).

Provide two solutions:
A) One using a for loop.
B) One using a while loop. (Do not use recursion)

- $1! = 1$
- $2! = 2 \times 1 = 2$
- $3! = 3 \times 2 \times 1 = 6$
- $4! = 4 \times 3 \times 2 \times 1 = 24$
- $5! = 5 \times 4 \times 3 \times 2 \times 1 = 120$

```python
def fac_for():
    r = 1
    n = int(input("Input your number: "))
    for i in range(1, n+1):
            r = r*i
    print(r)

fac_for()

def fac_while():
    i = 1
    r = 1
    n = int(input("Input your number: "))
    while(i <= n):
        r = r*i
        i += 1
        
    print(r)

fac_while()
```

Write `is_palindrome(word)` returning `True` if `word` reads the same backwards.  

```python
def is_palindrome(word):
    return word == word[::-1]

is_palindrome("ana")
```

### 2. Normalize
Write a function `normalize(xs)` that takes a list of numbers and returns a new list scaled to sum to 1.
- If the sum is 0, return a list of zeros of the same length.

```python
def normalize(xs):
    total = sum(xs)
    if total == 0:
        return [0 for _ in xs]
    return [x / total for x in xs]

normalize([1,-1])
```

### 3. Frequency table
Implement `freq(s)` returning a dict `{char: count}` for a string.

```python
def freq(s):
    counts = {}
    for char in s:
        if char in counts:
            counts[char] += 1
        else:
            counts[char] = 1
    return counts

freq("Eu gosto muito de python")
```

### 4. — List/dict comprehension
Given `records = [('Ana', 15), ('Bruno', 12), ('Carla', 18)]`, build:
1) a dict `grades`
2) a list of names with grade >= 14

```python
records = [('Ana', 15), ('Bruno', 12), ('Carla', 18)]

grades = {name: grade for name, grade in records}

top_students = [name for name, grade in records if grade >= 14]

print(grades)
print(top_students)
```

### 3. Try/Except
A) Write a function that converts a string to float, handling the case where the string is not a valid number.

B) Create a safe division function that handles division by zero.

```python
def safe_convert(text):
    try:
        return float(text)
    except ValueError:
        print(f"Error: '{text}' is not a valid number.")
        return None

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")
        return None

print(safe_convert("3.14"))
print(safe_divide(10, 2))
```

### 4. Recursion
A) Write a recursive function to calculate the power of a number: `power(base, exponent)`.

$$
2^4 = 2 \times 2 \times 2 \times 2 = 16
$$


B) Write a recursive function that counts down from a given number to 1, printing each number.

```python
def countdown(n):
    # Base Case: Stop when we hit 0 (or less)
    if n <= 0:
        print("Blastoff!")
        return

    # Action: Print the current number
    print(n)
    
    # Recursive Step: Call the function with the next number down
    countdown(n - 1)

# Usage
countdown(5)
# Output:
# 5
# 4
# 3
# 2
# 1
# Blastoff!
```
